```python
def gaussian_2d(x, y, amplitude, xo, yo, sigma_x, sigma_y):
    """
    Generates a 2D Gaussian function.
    """
    exponent = -((x - xo)**2 / (2 * sigma_x**2) + (y - yo)**2 / (2 * sigma_y**2))
    return amplitude * np.exp(exponent)

# Generate x and y coordinates
x, y = np.meshgrid(np.linspace(-7, 7, 100), np.linspace(-7, 7, 100))

# Create a 2D Gaussian
gaussian = gaussian_2d(x, y, 2, 0, 0, 2, 2)

fig = plt.figure()
ax = plt.axes(projection='3d')
# p1 = ax.plot_surface(x, y, gaussian, rstride=1, cstride=1,
#                 cmap=plt.cm.coolwarm, edgecolor='none', alpha=1.0)
p1 = ax.plot_surface(x, y, gaussian, rstride=1, cstride=1,
                cmap=mpl.colormaps["bwr"], edgecolor='none', alpha=1.0)
# p1 = ax.plot_surface(x, y, gaussian, rstride=1, cstride=1,
#                 cmap=mpl.colormaps["coolwarm"], edgecolor='none', alpha=1.0)
# ax.view_init(elev=30., azim=15)
plt.axis("off")
plt.show()
```

```python
def gaussian_2d(x, y, amplitude, xo, yo, sigma_x, sigma_y):
    """
    Generates a 2D Gaussian function.
    """
    exponent1 = -((x - xo)**2 / (2 * sigma_x**2) + (y - yo)**2 / (2 * sigma_y**2))
    exponent2 = -((x - 5*xo)**2 / (2 * sigma_x**2) + (y - 5*yo)**2 / (2 * sigma_y**2))
    return amplitude * (np.exp(exponent1) + np.exp(exponent2) )

# Generate x and y coordinates
x, y = np.meshgrid(np.linspace(-20, 20, 100), np.linspace(-20, 20, 100))

# Create a 2D Gaussian
gaussian = gaussian_2d(x, y, 2, 0, 0, 2, 2)

fig = plt.figure()
ax = plt.axes(projection='3d')
# p1 = ax.plot_surface(x, y, gaussian, rstride=1, cstride=1,
#                 cmap=plt.cm.coolwarm, edgecolor='none', alpha=1.0)
p1 = ax.plot_surface(x, y, gaussian, rstride=1, cstride=1,
                cmap=mpl.colormaps["bwr"], edgecolor='none', alpha=1.0)
# p1 = ax.plot_surface(x, y, gaussian, rstride=1, cstride=1,
#                 cmap=mpl.colormaps["coolwarm"], edgecolor='none', alpha=1.0)
# ax.view_init(elev=30., azim=15)
plt.axis("off")
plt.show()
```

```python
import plotly.graph_objects as go
import numpy as np

import plotly.io as pio
pio.renderers.default = 'iframe'

# Create a grid of x and y values
x = np.linspace(-10, 20, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)

# Calculate the Gaussian distribution
Z1 = np.exp(-(X**2 + Y**2) / 2 / 8) / 2 
Z2 = np.exp(-((X-5)**2 + (Y-5)**2) / 2 / 10) / 3
Z3 = np.exp(-((X-10)**2 + (Y+5)**2) / 2 / 6) / 2
Z = Z1 + Z2 + Z3

# Create the figure # Viridis
fig = go.Figure(data=[go.Surface(z=Z, x=x, y=y, colorscale='Bluered', colorbar=None)])

# Disable grid lines
fig.update_layout(
    scene=dict(
        xaxis=dict(showgrid=False, showticklabels=False, title=''),
        yaxis=dict(showgrid=False, showticklabels=False, title=''),
        zaxis=dict(showgrid=False, showticklabels=False, title='')
    )
)
fig.update_layout(
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)',
)
fig.update_scenes(xaxis_visible=False, yaxis_visible=False,zaxis_visible=False )
fig.update_traces(showscale=False)

# Default parameters which are used when `layout.scene.camera` is not provided
camera = dict(
    up=dict(x=0, y=0, z=1),
    center=dict(x=0, y=0, z=0),
    eye=dict(x=1.25, y=1.25, z=1.25)
)
fig.update_layout(scene_camera=camera)

# fig.write_image("2dmulti_gaussian.png")
# fig.write_html("2dmulti_gaussian.html")
fig.show()
```
